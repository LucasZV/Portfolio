# Data Cleaning


### 1. Merging diferent Datasets 

```python
import pandas as pd
```

```python
df_artists = pd.read_csv("../data/artists.csv")
df_spoty = pd.read_csv("../data/Spotify_dataset.csv")
df_features = pd.read_csv("../data/SpotifyFeatures.csv")
df_tracks = pd.read_csv("../data/tracks.csv")
```

```python
df_spoty.head()
```

```python
df_spoty.info()
```

```python
df_features.info()
```

DataFrame `df_features` has the same information as `df_tracks` therefore I won't use it.

```python
df_tracks.info()
```

```python
df_artists
```

```python
df_tracks = df_tracks.rename(columns= {"name" : "track_name"})
```

joining the datasets

```python
df_merge = pd.merge(df_spoty, df_tracks, on = "track_name", how = "left")
```

```python
df_merge.info()
```

```python
df_spotify = df_merge.copy()
```

```python
df_spotify.drop_duplicates("track_id", inplace= True, keep= "first")
```

```python
df_spotify.info()
```

cleaning the final dataframe

```python
df_spotify.drop("Unnamed: 0", axis= 1, inplace= True)
```

```python
df_spotify.duplicated().sum()
```

```python
df_spotify.drop(["id", "artists_y", "popularity_y", "duration_ms_y", "explicit_y", "artists_y", "danceability_y", "energy_y", "key_y", "loudness_y", "mode_y", "speechiness_y", "acousticness_y", "instrumentalness_y", "liveness_y", "valence_y", "tempo_y", "time_signature_y"], axis= 1, inplace = True)
```

```python
df_spotify.info()
```

```python
df_spotify.fillna("No_Value")
```

```python
df_spotify.rename(columns={
    "artists_x" : "artist",
    "popularity_x" : "popularity",
    "duration_ms_x" : "duration_ms",
    "explicit_x" : "explicit",
    "danceability_x" : "danceability",
    "energy_x" : "energy",
    "key_x" : "key",
    "loudness_x" : "loudness",
    "mode_x" : "mode",
    "speechiness_x" : "speechiness",
    "acousticness_x" : "acousticness",
    "instrumentalness_x" : "instrumentalness",
    "liveness_x" : "liveness",
    "valence_x" : "valence",
    "tempo_x" : "tempo",
    "time_signature_x" : "time_signature"
}, inplace= True)
```

```python
df_spotify.drop(columns= "id_artists", inplace= True)
```

```python
df_spotify.drop_duplicates("track_id", inplace= True, keep= "first")
```

```python
df_spotify.info()
```

```python
df_spotify
```

Extracting only the release year

```python
df_spotify["release_year"] = df_spotify["release_date"].str.extract(r'(\d{4})')
```

```python
df_spotify.drop(["release_date"], axis= 1, inplace= True)
```

```python
df_spotify
```

```python
df_spotify["release_year"] = pd.to_datetime(df_spotify["release_year"], format= "%Y")
df_spotify["release_year"] = df_spotify["release_year"].dt.year
```

```python
df_spotify
```

short by popularity

```python
df_spotify = df_spotify.sort_values(by = "popularity", ascending= False)
```

Transform ``duration`` from ms to minutes

```python
df_spotify["duration_ms"] = round(df_spotify["duration_ms"] / 60000, 2)
df_spotify.rename(columns= {"duration_ms" : "duration_mins"}, inplace = True)
```

```python
df_spotify
```

```python
df_spotify_def = df_spotify.copy()
```

Saving the inal datarame to .csv and check.

```python
df_spotify_def.to_csv("../data/df_spotify_clean.csv", index= False)
```
